# Data-Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('MSFT.csv')
data.head()
```

```python
del data['Date']
del data['Adj Close']
```

```python
data.columns
```

```python
plt.figure(figsize=(20,10))
plt.scatter(y='Close',x='Open',data=data)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='High',data=data)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.hist(data['Open'],color='red')

plt.subplot(1,2,2)
plt.hist(data['High'],color='green')
```

```python
data.hist(figsize=(20,60),layout=(15,4),color='green')
plt.show()
```

```python
plt.boxplot(data['Close'])
```

```python
sns.displot(data['High'],color='purple')
```

```python
sns.pairplot(data)
```

```python
fig,ax=plt.subplots(figsize=(30,20))
sns.heatmap(data.corr(),annot=True,fmt='0.2%',cmap='autumn',ax=ax)
```
